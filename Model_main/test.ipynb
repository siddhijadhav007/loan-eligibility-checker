```python
import requests

url = "http://127.0.0.1:8000/predict"

data = {
    "Married": 1.0,
    "Education": 1.0,
    "Self_Employed": 0.0,
    "ApplicantIncome": 5703.0,
    "CoapplicantIncome": 0.0,
    "LoanAmount": 128.0,
    "Credit_History": 1.0,
    "Property_Area": 2.0
}

try:
    response = requests.post(url, json=data)
    print("Status Code:", response.status_code)
    print("Response:", response.text)
except Exception as e:
    print("Error:", e)
```
